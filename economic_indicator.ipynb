```python
def json_to_df(data, column_name):
  assert isinstance(column_name, str)
  df = pd.DataFrame(data['data'])
  df['date'] = pd.to_datetime(df['date'])
  df.set_index('date', inplace=True)
  df['value'] = pd.to_numeric(df['value'])
  df.rename(columns={'value': column_name}, inplace=True)
  return df
```

### Annual GDP

```python
url = f'https://www.alphavantage.co/query?function=REAL_GDP&interval=annual&apikey={api_key}'
r = requests.get(url)
gdp_data = r.json()
```

```python
gdp = json_to_df(gdp_data, 'annual_gdp')
gdp = gdp.resample('MS').ffill()
gdp
```

```python
gdp = gdp.resample('1D').ffill()
gdp
```

```python
dfs = [gdp, treasury_yield_10yr, treasury_yield_5yr, treasury_yield_3month, federal_fund_rate, cpi, inflation_rate, retail_sales, durable_goods, unemployment, nonfarm_payroll]
merged = pd.concat(dfs, join='outer', axis=1)
merged = merged.dropna()
merged
```

```python
import pandas as pd

# Calculate percentage changes for each economic indicator
for column in merged.columns:
    if column != 'date':
        merged[f'{column}_pct_change'] = merged[column].pct_change() * 100

# Calculate rise/decline indicators for each indicator
for column in merged.columns:
    if column != 'date' and not column.endswith('_pct_change') and not column.startswith('buy_signal_') and not column.startswith('sell_signal_'):
        merged[f'rise_decline_{column}'] = merged[column].diff().apply(lambda x: 1 if x > 0 else -1 if x < 0 else 0)

# Drop the first row as it will have NaN values due to percentage change calculation
merged = merged.dropna()

# Drop the original raw data columns
columns_to_drop = [column for column in merged.columns if not column.endswith('_pct_change') and not column.startswith('rise_decline_') and column != 'date']
merged = merged.drop(columns=columns_to_drop)

# Display the resulting DataFrame
print(merged.head())
```
